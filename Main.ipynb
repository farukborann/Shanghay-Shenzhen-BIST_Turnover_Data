```python
import akshare as ak
import pandas as pd
import yfinance as yf
from datetime import datetime
```

```python
class ExcelFile:
    writer = None
    def __init__(self, fp):
        global writer
        writer = pd.ExcelWriter(fp, engine='xlsxwriter')
        
    def CreateNewSheetWithData(self, df, sn, index=True):
        df.to_excel(writer, sheet_name=sn, index=index)

    def Close(self):
        writer.save()
        writer.close()
```

```python
end_date_yahoo = datetime.now().strftime("%Y-%m-%d")

USDCNY_data = yf.download('USDCNY=X', start='2000-01-01', end=end_date_yahoo)
USDCNY_data["AVG"] = (USDCNY_data["Open"] + USDCNY_data["High"] + USDCNY_data["Low"] + USDCNY_data["Close"]) / 4
USDCNY_data.reset_index(inplace=True)
USDCNY_data['Date'] = pd.to_datetime(USDCNY_data['Date'])
```

```python
end_date = datetime.now().strftime("%Y%m%d")
start_date = USDCNY_data.iloc[0]["Date"].strftime("%Y%m%d")
```

```python
def CNY2USD(cny, date, dec = 2):
    dates = USDCNY_data[USDCNY_data["Date"] == date]
    if(len(dates) > 0):
        return round(cny * dates.iloc[0]["AVG"], dec)
    else:
        # print("Date Not Founded : ", date)
        return -1
```

```python
shanghay_composite_index = ak.stock_zh_a_hist_163(symbol="sh000001", start_date=start_date, end_date=end_date)
shanghay_composite_index.columns = ["Date", "Stock Code", "Name", "Closing Price", "Highest Price", "Lowest Price", "Open Price", "Previous Close", "Amount up or down", "Up/Down",
"Change hands rate", "Volume", "Turnover", "Total Market Capitalization", "Market Capitalization"]
shanghay_composite_index = shanghay_composite_index.drop(["Name", "Stock Code"], axis=1)
```

```python
shanghay_composite_index["Turnover (USD)"] = shanghay_composite_index.apply(lambda row: CNY2USD(row['Turnover'], row['Date']), axis=1)
shanghay_composite_index = shanghay_composite_index[shanghay_composite_index["Turnover (USD)"] != -1]
```

```python
shanghay_composite_index['Date'] = pd.to_datetime(shanghay_composite_index['Date'])
shanghay_composite_index
```

```python
shenzhen_composite_index = ak.stock_zh_a_hist_163(symbol="zs399106", start_date="20000101", end_date='20221125')
shenzhen_composite_index.columns = ["Date", "Stock Code", "Name", "Closing Price", "Highest Price", "Lowest Price", "Open Price", "Previous Close", "Amount up or down", "Up/Down",
"Change hands rate", "Volume", "Turnover", "Total Market Capitalization", "Market Capitalization"]
shenzhen_composite_index = shenzhen_composite_index.drop(["Name", "Stock Code"], axis=1)
shenzhen_composite_index['Date'] = pd.to_datetime(shenzhen_composite_index['Date'])
```
